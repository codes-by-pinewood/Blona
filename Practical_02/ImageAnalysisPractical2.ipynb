### **4: Graphs**



---


Here we plot the square root of the mean (on the x-axis) of the **area** of objects against the coefficient of variation: CV = $\frac{ σ}{ μ'}$ (on the y-axis). For this we employed numpy libraries like ``` np.mean ``` and ``` np.std```
.

````python
import matplotlib.pyplot as plt

area_A = np.array([0.0001285])
area_B = np.array([3195, 3285 ,3265,3248, 3230, 3262])
area_C = np.array([1320, 1328,1335, 1304, 1340 ,1309, 1297, 1360])
area_D = np.array([475.0 , 480.0, 480.0 , 480.0 , 486.0, 473.0, 468.0, 499.0 , 484.0 , 505.0])

mean_A = np.mean(area_A)
mean_B = np.mean(area_B)
mean_C = np.mean(area_C)
mean_D = np.mean(area_D)

sqrt_mean_A = np.sqrt(mean_A)
sqrt_mean_B = np.sqrt(mean_B)
sqrt_mean_C = np.sqrt(mean_C)
sqrt_mean_D = np.sqrt(mean_D)


cv_a = np.std(area_A)/mean_A
cv_b = np.std(area_B)/mean_B
cv_c = np.std(area_C)/mean_C
cv_d = np.std(area_D)/mean_D

sqrt_mean_array = [sqrt_mean_A,sqrt_mean_B,sqrt_mean_C,sqrt_mean_D]

cv_array = [cv_a,cv_b,cv_c,cv_d]

#print(sqrt_mean_array)
#print(cv_array)

z = ['Rect A', 'Rect B', 'Rect C', 'Rect D']

fig, ax = plt.subplots()
ax.scatter(sqrt_mean_array, cv_array)

for i, txt in enumerate(z):
    ax.annotate(txt, (sqrt_mean_array[i], cv_array[i] ))
````

Here we plot the square root of the mean (on the x-axis) of the **perimeter** of objects against the coefficient of variation: CV = $\frac{ σ}{ μ'}$ (on the y-axis). For this we employed numpy libraries like ``` np.mean ``` and ``` np.std```
.

````python

perimeter_A = np.array([457.5])
perimeter_B = np.array([233.7, 230.1, 234.7, 236.0, 233.3, 235.1])
perimeter_C = np.array([143.9, 148.8, 148.9, 146.9, 147.0, 147.5, 147.2, 150.5])
perimeter_D = np.array([92.43,90.98, 90.98, 92.92, 93.35, 92.05, 92.67, 93.54, 94.08, 95.23])

mean_pA = np.mean(perimeter_A)
mean_pB = np.mean(perimeter_B)
mean_pC = np.mean(perimeter_C)
mean_pD = np.mean(perimeter_D)

mean_array = np.array([mean_pA, mean_pB, mean_pC, mean_pD])


cv_a = np.std(perimeter_A)/mean_pA
cv_b = np.std(perimeter_B)/mean_pB
cv_c = np.std(perimeter_C)/mean_pC
cv_d = np.std(perimeter_D)/mean_pD

sqrt_mean_array = np.sqrt(mean_array)
cv_array = np.array([cv_a, cv_b, cv_c, cv_d])


z = ['Rect A', 'Rect B', 'Rect C', 'Rect D']

fig, ax = plt.subplots()
ax.scatter(sqrt_mean_array, cv_array)

for i, txt in enumerate(z):
    ax.annotate(txt, (sqrt_mean_array[i], cv_array[i]))
````
